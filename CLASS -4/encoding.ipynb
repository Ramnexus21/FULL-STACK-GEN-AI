```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
documents =[
    "people watch movies",
    "people watch cricket",
    "people like movies",
    "people like cricket"
]
```

##list comprehension##

```python
tokens = [sentence.lower().split() for sentence in documents]
```

```python
tokens
```

```python
all_words = [[word] for sentence in tokens for word in sentence]
```

```python
all_words
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoder.fit (all_words)
encoder.categories_[0]
```

```python
print("vocabulary: ", encoder.categories_[0])
```

```python
for sentence in tokens:
    print(sentence)
```

```python
for sentence in tokens:
    encoded_sentence = encoder.transform([[word] for word in sentence])
    encoded_sentence = encoded_sentence.astype(int)
    print("\nEncoded sentence:", sentence)
    print("Encoded:", encoded_sentence)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
bow = CountVectorizer()
bow_matrix = bow.fit_transform(documents)
```

```python
bow_matrix
```

```python
bow_matrix.toarray()
```

```python
bow.get_feature_names_out()
```

```python
print("vocabulary:\n", bow.get_feature_names_out())
```
